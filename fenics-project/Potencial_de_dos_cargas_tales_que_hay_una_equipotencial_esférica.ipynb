```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as mpl

#situamos los ejes coordenados de forma que las cargas estan sobre el eje x

a = 1.
r = 2.*a
rp = a*a/r

x_vals = np.linspace(-a*1.2, a*2.8, 200)
y_vals = np.linspace(-1.5*a, 1.5*a, 200)
X,Y = np.meshgrid(x_vals,y_vals)


#definimos el potencial para z=0

Z= 1./np.sqrt((X-r)**2+Y**2)  -  rp/a/np.sqrt((X-rp)**2+Y**2)


# Ahora graficamos el potencial 

fig = mpl.figure(figsize=(9,10))

levels = np.arange(-14., 14., 2)

cp0= mpl.contourf(X,Y,Z,levels)

cp1= mpl.contour(X,Y,Z,levels,linewidths=1.5,colors='k')

mpl.title('Two particles')
mpl.xlabel('x')
mpl.ylabel('y')

fig.colorbar(cp0,orientation="horizontal")
mpl.show()

```
